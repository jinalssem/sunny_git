```python
def 암호체크(암호):  # 이 함수는 위의 암호에 중복이 있는지 체크하는 역할을 한다.
    counts=dict()
    for w in 암호:
        counts[w]=counts.get(w,0)+1
    for k in counts:
        if counts[k]>1:
            print(k,end=" ")

def 애니그마(문장, 암호):    # 암호화
  encoding=''
  for w in 문장:
    if w.lower() in 암호:
      if w.isupper():
        encoding = encoding + 암호[(암호.index(w.lower())+1)%len(암호)].upper()
      else:
        encoding = encoding + 암호[(암호.index(w.lower())+1)%len(암호)]
    else:
      encoding = encoding + w
  return encoding

def 앨런튜링(문장, 암호):     # 복호화
  decoding=''
  for w in 문장:
    if w.lower() in 암호:
      if w.isupper():
        decoding = decoding + 암호[(암호.index(w.lower())-1)%len(암호)].upper()
      else:
        decoding = decoding + 암호[(암호.index(w.lower())-1)%len(암호)]
    else:
      decoding = decoding + w
  return decoding

# 파일 업로드
def 파일_처리(파일명, 암호):
    with open(파일명, 'r', encoding='utf-8') as f:
        file_content = f.read()
        print('Uploaded file:', 파일명, '\n')
        print("평문: ")
        print(file_content)
        encoded = 애니그마(file_content,암호)        # 암호화
        print("암호문: ")
        print(encoded)
        decoded = 앨런튜링(encoded, 암호)     # 복호화
        print("복호문: ")
        print(decoded)


from google.colab import files
암호="국및광도송역별억묶평균 '색홍되학며신삽량제통계더49awsxdecrvfbgtyhnuljmkiop05얼활정동굴연~료대결녀종객의쌤체넣작따길옴앨서표큰머닥형걸치뿌클혁성쌍파썬럼두처쑤번니용확개닭남습변뒤있환때리트를추출면양헌철규상워실레쓰기한와셔애님움던절끓는키석피젊진들배문담입할력오게포말전올런녕생려라돌튜토밥었나?비!점밤불없달둬영샵까만로중씨혹식옳것겁년잘된풍져으벌음손크반다랑은질내수조재선근새임스링갈심위타래류이세김사줘순최미지항희황못준명화우택시버윤에보바박좋해호가암구저을끼침그녁인마뿔줌탄씹차야멍산천고주아빨먹장싶좆어현소꼬자똥안하요액"

if __name__=='__main__':
    암호체크(암호)      # 중복된 글자가 있는지 확인

    # 파일을 업로드합니다.
    uploaded = files.upload()

    # 업로드된 파일의 이름을 출력합니다.
    for filename in uploaded.keys():
        파일_처리(filename, 암호)
```
